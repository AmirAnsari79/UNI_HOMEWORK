```python
read = ''

with open('romeo.txt') as f:
    for _ in range(200):
        read += f.readline()
read = read.strip()
print(read)
```

```python
read = read.split('\n')
read = [i.lower() for i in read if len(i) != 0]
read = [i.strip() for i in read]
read = '\n'.join(read)
print(read)  #<str>
```

```python
d = dict()
for char in read:
    if not char.isalpha():
        read = read.replace(char, ' ')

read = read.strip(' ')
read = read.split()

for item in read:
    d[item] = d.get(item, 0) + 1
print(d)
```

```python
read=''
new_word=''
with open('romeo.txt') as f:
    for _ in range(100):
        read += f.readline()
read = read.strip()
read=read.split()
for chng in read:
    new_word+=chng.capitalize()
    new_word+=' '
print(new_word)
with open('new_text.txt','w') as f:
    f.write(new_word)
```
